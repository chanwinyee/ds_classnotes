```python
tot_cols = [col for col in df.columns if col.startswith("Totals")]
rates_cols = [col for col in df.columns if col.startswith("Rates")]
rates_df = df.drop(columns=tot_cols)
tot_df = df.drop(columns=rates_cols)
```

```python
rates_alcohol_cols = [col for col in df.columns if col.startswith("Rates.Alcohol")]
rates_illicit_cols = [col for col in df.columns if col.startswith("Rates.Illicit Drugs")]
rates_marijuana_cols = [col for col in df.columns if col.startswith("Rates.Marijuana")]
rates_pain_relievers_cols = [col for col in df.columns if col.startswith("Rates.Pain Relievers")]
rates_tobacco_cols = [col for col in df.columns if col.startswith("Rates.Tobacco")]


rates_alcohol_df = df.drop(columns=rates_alcohol_cols)
rates_illicit_df = df.drop(columns=rates_illicit_cols)
rates_marijuana_df = df.drop(columns=rates_marijuana_cols)
rates_pain_relievers_df = df.drop(columns=rates_pain_relievers_cols)
rates_tobacco_df = df.drop(columns=rates_tobacco_cols)

```
